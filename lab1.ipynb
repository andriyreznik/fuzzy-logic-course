```python
import matplotlib.pyplot as plt
import numpy as np
```

Статистичний метод обробки експертної інформації.

Варіант №7:

Кількість експертів - 6; <br>
Універсальна множина, X - 14, 22, 31, 64, 82; <br>
Нечітка змінна, A - якість продукції (%); <br>
Лінгвістичні терми - низька, середня, висока. <br>

```python
m = 6
x = [14, 22, 31, 64, 82]
```

Задаємо матрицю попарних порівнянь:

```python
Q = np.array([[6, 4, 3, 0, 0],
              [0, 2, 3, 4, 1],
              [0, 0, 0, 2, 5]])
```

Обчислюємо значення функції належності нечіткої множини:

```python
mu =  ((1 / m) * Q).round(2)
print(mu.round(2))
```

Записуємо нечіткі множини, які характеризують якість продукції:

$A_{низька}$ = $\frac{1}{14} + \frac{0.67}{22} + \frac{0.5}{31}$

$A_{середня}$ = $\frac{0.33}{22} + \frac{0.5}{31} + \frac{0.67}{64} + \frac{0.17}{82}$

$A_{висока}$ = $\frac{0.33}{64} + \frac{0.83}{82}$

Будуємо графік функцій належності термів, які характеризують якість продукції (%):

```python
plt.plot(x, mu[0], 'y', marker="o", label="низька")
plt.plot(x, mu[1], 'm-', marker="o", label="середня")
plt.plot(x, mu[2], 'c:', marker="o", label="висока")
plt.xticks(x)
plt.yticks(np.unique(mu))
plt.legend(loc="best")
plt.show()
```
